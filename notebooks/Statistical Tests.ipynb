```python
from scipy.stats import *

# naive resampling
krus_acc_stat_naive, krus_acc_p_naive = kruskal(naive_CNN_acc, naive_RFC_acc)
krus_pre_stat_naive, krus_pre_p_naive = kruskal(naive_CNN_pre, naive_RFC_pre)
krus_rec_stat_naive, krus_rec_p_naive = kruskal(naive_CNN_rec, naive_RFC_rec)
krus_f1s_stat_naive, krus_f1s_p_naive = kruskal(naive_CNN_f1s, naive_RFC_f1s)

# smote
krus_acc_stat_smote, krus_acc_p_smote = kruskal(smote_CNN_acc, smote_RFC_acc)
krus_pre_stat_smote, krus_pre_p_smote = kruskal(smote_CNN_pre, smote_RFC_pre)
krus_rec_stat_smote, krus_rec_p_smote = kruskal(smote_CNN_rec, smote_RFC_rec)
krus_f1s_stat_smote, krus_f1s_p_smote = kruskal(smote_CNN_f1s, smote_RFC_f1s)

# cnn synth
krus_acc_stat_cnn_synth, krus_acc_p_cnn_synth = kruskal(cnn_synth_CNN_acc, cnn_synth_RFC_acc)
krus_pre_stat_cnn_synth, krus_pre_p_cnn_synth = kruskal(cnn_synth_CNN_pre, cnn_synth_RFC_pre)
krus_rec_stat_cnn_synth, krus_rec_p_cnn_synth = kruskal(cnn_synth_CNN_rec, cnn_synth_RFC_rec)
krus_f1s_stat_cnn_synth, krus_f1s_p_cnn_synth = kruskal(cnn_synth_CNN_f1s, cnn_synth_RFC_f1s)

# cnn sparse synth
krus_acc_stat_cnn_sparse_synth, krus_acc_p_cnn_sparse_synth = kruskal(cnn_sparse_synth_CNN_acc, cnn_sparse_synth_RFC_acc)
krus_pre_stat_cnn_sparse_synth, krus_pre_p_cnn_sparse_synth = kruskal(cnn_sparse_synth_CNN_pre, cnn_sparse_synth_RFC_pre)
krus_rec_stat_cnn_sparse_synth, krus_rec_p_cnn_sparse_synth = kruskal(cnn_sparse_synth_CNN_rec, cnn_sparse_synth_RFC_rec)
krus_f1s_stat_cnn_sparse_synth, krus_f1s_p_cnn_sparse_synth = kruskal(cnn_sparse_synth_CNN_f1s, cnn_sparse_synth_RFC_f1s)

# lstm synth
krus_acc_stat_lstm_synth, krus_acc_p_lstm_synth = kruskal(lstm_synth_CNN_acc, lstm_synth_RFC_acc)
krus_pre_stat_lstm_synth, krus_pre_p_lstm_synth = kruskal(lstm_synth_CNN_pre, lstm_synth_RFC_pre)
krus_rec_stat_lstm_synth, krus_rec_p_lstm_synth = kruskal(lstm_synth_CNN_rec, lstm_synth_RFC_rec)
krus_f1s_stat_lstm_synth, krus_f1s_p_lstm_synth = kruskal(lstm_synth_CNN_f1s, lstm_synth_RFC_f1s)

# lstm sparse synth
krus_acc_stat_lstm_sparse_synth, krus_acc_p_lstm_sparse_synth = kruskal(lstm_sparse_synth_CNN_acc, lstm_sparse_synth_RFC_acc)
krus_pre_stat_lstm_sparse_synth, krus_pre_p_lstm_sparse_synth = kruskal(lstm_sparse_synth_CNN_pre, lstm_sparse_synth_RFC_pre)
krus_rec_stat_lstm_sparse_synth, krus_rec_p_lstm_sparse_synth = kruskal(lstm_sparse_synth_CNN_rec, lstm_sparse_synth_RFC_rec)
krus_f1s_stat_lstm_sparse_synth, krus_f1s_p_lstm_sparse_synth = kruskal(lstm_sparse_synth_CNN_f1s, lstm_sparse_synth_RFC_f1s)
```

## Post Hoc Nemenyi Test

```python
acc_data = np.array([naive_CNN_acc.to_numpy(), smote_CNN_acc.to_numpy(), cnn_synth_CNN_acc.to_numpy(), cnn_sparse_synth_CNN_acc.to_numpy(), lstm_synth_CNN_acc.to_numpy(), lstm_sparse_synth_CNN_acc.to_numpy()])
acc_data = np.squeeze(acc_data).T

pre_data = np.array([naive_CNN_pre.to_numpy(), smote_CNN_pre.to_numpy(), cnn_synth_CNN_pre.to_numpy(), cnn_sparse_synth_CNN_pre.to_numpy(), lstm_synth_CNN_pre.to_numpy(), lstm_sparse_synth_CNN_pre.to_numpy()])
pre_data = np.squeeze(pre_data).T

rec_data = np.array([naive_CNN_rec.to_numpy(), smote_CNN_rec.to_numpy(), cnn_synth_CNN_rec.to_numpy(), cnn_sparse_synth_CNN_rec.to_numpy(), lstm_synth_CNN_rec.to_numpy(), lstm_sparse_synth_CNN_rec.to_numpy()])
rec_data = np.squeeze(rec_data).T

f1s_data = np.array([naive_CNN_f1s.to_numpy(), smote_CNN_f1s.to_numpy(), cnn_synth_CNN_f1s.to_numpy(), cnn_sparse_synth_CNN_f1s.to_numpy(), lstm_synth_CNN_f1s.to_numpy(), lstm_sparse_synth_CNN_f1s.to_numpy()])
f1s_data = np.squeeze(f1s_data).T
```

```python
from scikit_posthocs import posthoc_nemenyi_friedman as ph_nf
acc_nemenyi_cnn = ph_nf(acc_data)
pre_nemenyi_cnn = ph_nf(pre_data)
rec_nemenyi_cnn = ph_nf(rec_data)
f1s_nemenyi_cnn = ph_nf(f1s_data)
```
